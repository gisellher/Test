```python
#Importar las librerias
import requests
import pandas as pd
import datetime as dt
import io
```

EXTRACCIÓN

```python
#Traer datos
urlData = requests.get('https://datos.gov.co/resource/7cci-nqqb.json')
#Convertir la respuesta a formato JSON
jsonData = urlData.json()
#Crear un DataFrame a partir de los datos JSON
data = pd.DataFrame(jsonData)
pd.set_option('display.max_columns', None)
# Ahora 'df' contiene los datos en forma de DataFrame
#print(data)
data
```

LIMPIEZA

```python
#Eliminar columnas de año, mes y día ya que ya tenemos una columna fecha
data = data.drop(['a_o', 'mes', 'd_a'], axis=1)
#Renombrar la columna "campaero" por estar mal escrita
data.rename(columns={'campaero':'campero'}, inplace=True)
#Datos
data
```

```python
#Manejo de columna fecha y hora
#dt.datetime.strptime(data['fecha'], "%d/%m/%Y")
```

```python
data = data.astype({'orden': 'int', 'fecha': 'object', 'peaton': 'int', 'automovil': 'int', 'campero': 'int', 'camioneta': 'int', 'micro': 'int', 'buseta': 'int', 'bus': 'int', 'camion': 'int', 'volqueta': 'int', 'moto': 'int', 'bicicleta': 'int', 'otro': 'int',  })
data.dtypes
```

TRANSFORMACIÓN

```python
#Para llevar un orden a los datos, usaremos Orden como valor de guia

guia = 'orden'

#Divir en cuatro DataFrames, cada uno con columnas separadas

principal = data[['fecha', 'gravedad', 'entidad', 'propietario_de_veh_culo', guia]].copy()
agente_vial = data[['peaton', 'automovil', 'campero', 'camioneta', 'micro', 'buseta', 'bus', 'camion', 'volqueta', 'moto', 'bicicleta', 'otro', guia]].copy()
lugar = data[['via_1', 'barrio', 'nombrecomuna', guia]].copy()
tiempo = data[['hora', 'diurnio_nocturno', 'hora_restriccion_moto']].copy()

print("Principal:")
print(principal)

print("\nAgente vial:")
print(agente_vial)

print("\nLugar:")
print(lugar)

print("\nTiempo:")
print(tiempo)
```
